```python
# Thực hành vẽ mặt cười
import turtle
pen = turtle.Turtle()
pen.pensize(5)
pen.fillcolor("#FFF700")
pen.shape("square")

pen.penup()
pen.goto(0, -200)

facesize = 200
eye_size = 30
nose_size = 20
# Vẽ mặt
pen.pendown()
pen.begin_fill()
pen.fillcolor("#184D02")
pen.circle(facesize)
pen.end_fill()

pen.penup()
pen.goto(-75, 75)

# draw an eye
pen.pendown()
pen.begin_fill()
pen.fillcolor("Red")
pen.circle(eye_size)
pen.end_fill()

pen.penup()
pen.goto(75, 75)

# draw an eye
pen.pendown()
pen.begin_fill()
pen.fillcolor("Blue")
pen.circle(eye_size)
pen.end_fill()

# Vẽ mũi
pen.penup()
pen.goto(0, 0)

pen.pendown()
pen.begin_fill()
pen.fillcolor("#FFFFFF")
pen.circle(nose_size, steps=3)
pen.end_fill()

# Vẽ mồm
pen.penup()
pen.goto(-100, -50)

pen.pendown()
pen.begin_fill()
pen.fillcolor("#FFF700")
pen.right(90)
pen.circle(100,180)
pen.end_fill()




turtle.done()
```
